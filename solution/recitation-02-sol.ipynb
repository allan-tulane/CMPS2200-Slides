5. We have $W(n) = a*W(n/b)+n^c$

Root cost: $n^c$

Level 1 cost: $a*(n/b)^c$

We just need to calculate the limit of $\lim_{n=\infty}n^c/(a*(n/b)^c) = b^c/a$

* $b^c > a$, i.e., $c > \log_ba$: leaf-dominated
* $b^c < a$, i.e., $c < \log_ba$: root-dominated
* $b^c = a$, i.e., $c = \log_ba$: balanced
